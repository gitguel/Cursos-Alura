# Matplotlib

```
notas = []

for i in range(8):
  notas.append(randrange(0,11))
```

```
from random import randrange
import matplotlib.pyplot as plt

x = list(range(1,9))
y = notas

print(f"Notas de Matemática: {y}\n")
plt.plot(x,y,marker="s")
plt.title("Notas de Matemática",)
plt.xlabel("Provas")
plt.ylabel("Notas")
plt.show()
```
